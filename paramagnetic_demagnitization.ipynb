# Adiabatic Demagnatization of Paramagnetic Salt 

Adiabatic Demagnetization uses the paramagnetic nature in some materials to cool those materials, generally in the form of gaseous, into the colder range or millikelvin. Here, I have simulated the demagnitization for $Gadolinium$. 

The final equation for final tempereture due to demagnatization of paramagnetic salt is given by,
$$T_2 = T_1 \left[ 1 - \dfrac{K}{C_H}.\dfrac{H}{T_1}\right]^{\dfrac{1}{2}}$$
Where ,
\begin{align}
T_1 &= Initial \ Temperature \\
K &= Curie \ Constant \\
C_H &= Curie-Weiss \ Constant \\ 
H &= Magnetic \ Field \ Strength \\
\end{align}

The constants for $Gadolinium$ are given as,

\begin{align}
K &=  0.1 K/Tesla \\
C_H &= 13 K \\
\end{align}

For $Gadolinium$ we are varying paramaters like $H$ from $0$ to few $Tesla$. Similarly we are varying value of $T_1$ from range $0.05$ to $10 K $.

## Programming a Equation

Let us firstly import modules which will make the task easier.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Now let us define a function $final\_tempr$ which takes two variable values $H$ and $T_1$ as a input, process the equation and executes the result.

```python
def final_tempr(H,T1):
    K = 0.1
    CH = 293
    T_2 = T1 *((1 - (K * H)/(CH * T1))**0.5)
    return T_2
```

We defined the function which executes the final temperature after taking two values $H$ and $T1$ as arguments. \\[5pt]

Now it's time to make two variables within certain range, i.e. $0.1$ to $5$ $Tesla$ for $H$ and $0.05$ to $10$ $K$ for $T_1$.


```python
H = np.linspace(0.1,5,100)
T1 = np.linspace (0.05, 10, 100)
T1
```

Now it's time to implement the variables into the function and get the final tempreture as a result. Following code gives the array of $T_2$ of size $100$.

```python
final_tempr(H,T1)
```

Since we got the array of Tempreture, we can now plot and interpreate the result.

```python
plt.grid("science",color="green")
plt.xlabel("Varying both $H$ and $T$")
plt.ylabel("Final Temperature ($T_2$)")
plt.plot(final_tempr(H,T1))
```

Since, we got the plot for both varying value. Let's just make either $H$ or $T$ constant to interpreate the result.

```python
H = 5
T1 = np.linspace (0.05, 10, 100)
```

Now let's plot the result.

```python
plt.grid("science",color="green")
plt.xlabel("Varying $T$ at constant $H = 5 T$")
plt.ylabel("Final Temperature ($T_2$)")
plt.plot(final_tempr(5,T1),color="red")
```

Now lets make Temperture constant, $T_1 = 5 K $ and vary $H$.

```python
T1 = 5
H = np.linspace (0.05, 10, 100)
```

```python
plt.grid("science",color="green")
plt.xlabel("Varying $H$ at constant $T = 5 K$")
plt.ylabel("Final Temperature ($T_2$)")
plt.plot(final_tempr(H,T1))
```
